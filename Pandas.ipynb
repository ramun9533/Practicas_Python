## Info

```python
df2.info()
```

```python
countries_dict = {
    "pais": ["Argentina", "Bélgica", "Finlandia", "Chile", "México"],
    "capital": ["Buenos Aires", "Brusela", "Helsinki", "Santiago", "CDMX"],
    'continente': ['América', 'Europa', 'Europa', 'América', 'América'],
    "area": [8.516, 17.10, 3.286, 9.597, 1.221],
    "poblacion": [200.4, 143.5, 125.2, 135.7, 52.98]
    }

countries_dict
```

```python
df = pd.DataFrame(countries_dict)
```

```python
df
```

## columns

```python
df.columns
```

## index

```python
df.index
```

```python
# Agregando un índice personalizado
df.index = ["AR", "BE", "FI", "CH", "MX"]
```

```python
df
```

```python
# Usando corchetes simples para obtener una columna
# ó un objeto de tipo Series
type(df["pais"])
```

```python
# usando corchetes dobles para obtener un
# dataframe
type(df[["pais"]])

# df[["pais"]] = pd.DataFrame(df["pais"])

# Nota: Es importante recordar esta sintanxis, por que,
# hay métodos que solo pertenecen a un DataFrame o a un Series
```

```python
# Creando una nueva columna
df["densidad_poblacion"] = df["poblacion"] / df["area"]
```

## round

```python
# Redondeando densidad de población
df["densidad_poblacion"] = df["densidad_poblacion"].round(2)
```

```python
df
```

## head

```python
# Regresa las primeras n filas
# Por default n = 5

df.head(3)
```

## sample

```python
df.sample(3)
```

# loc

Nos permite acceder a elementos de una lista, por etiquetas

```python
df.loc["BE"] # df.loc[1]
```

```python
df.loc[["AR", "MX"]]
```

```python
df.loc[["AR", "MX"], ["pais", "area", "capital"]]
```

## iloc
Nos permite acceder a elementos de una lista por posición númerica o index

```python
df.iloc[2]
```

```python
df.iloc[[0, 4]]
```

## describe

Es un método que nos regresa estadísticos descriptivos, relacionados con:
- tendencia central
- dispersión
- forma de la tabla

```python
df.describe(include="all")
```

## plot

```python
df.plot(x="pais", y="densidad_poblacion", kind="bar")
```
