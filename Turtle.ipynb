```python
initializeTurtle(initial_speed=7)
colors_list = ["red", "purple", "blue", "green", "orange", "yellow"]
for value in range(360):
  pencolor(colors_list[value%6])
  width(int(value/100) + 1)
  forward(value/1.5)
  left(59) #permite que la tortuga gire a la izquierda recibe enteros
```

```python
grados = 360
lados = int(input('cuantos lados quieres tu figura: '))
lados_grados = grados / lados
largor = int(input('largo quieres cada lado tu figura: '))
initializeTurtle(initial_speed=8)
colors_list = ["red", "purple", "blue", "green", "orange", "yellow"]
#left(grados/2)
#forward(250)
#pencolor("purple")
for i in range(lados):
  pencolor(colors_list[i%6])
  forward(largor)
  left(lados_grados)
  for i in range(lados):
    #pencolor(colors_list[i%6])
    forward(largor / 2)
    left(lados_grados)
    for i in range(lados):
    #pencolor(colors_list[i%6])
      forward(largor / 4)
      left(lados_grados)
```
